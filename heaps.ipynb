**HEAP**

Es una fila con prioridad, los datos se van a formar por cuál es el más importante. (Adelante va el más importante).

- El montículo(heap) es un árbol binario en el que los padres tienen mayor prioridad que los hijos.
- Siempre es un árbol completo. (Siempre están llenos todos los niveles excepto el último que se está llenando).


*Cómo puedo obtener el más importante?*

1. Se intercambia el primer nodo del árbol (raíz) por el último. (No pasa nada porque no tiene hijos)
2. Eliminar el último elemento
3. Hundir la raíz a su posición: Intercambiar la raíz con el hijo que es más grande (hasta que quede ordenado)

*Cómo meto un valor?*
1. Inserto el valor como último elemento
2. Flotar el elemento (comparar con su padre hasta que sea menor que él) (Se puede hacer de forma recursiva o iterativa)

Siempre se tarda O(Log(n)) en cualquier caso, algoritmo super estable.

**¿Como se implementa en memoria?**


- Se ignora la posición 0
- Los hijos de cada uno de los nodos en el arreglo son: 2*n izquierda y 2*n+1 derecha (hijos de 1 son 2 y 3)


**Heap sort**: Se meten los elementos de un arreglo a un montículo y se vuelven a sacar, quedando así ordenados. Tiene complejidad siempre de O(nLog(n)).
